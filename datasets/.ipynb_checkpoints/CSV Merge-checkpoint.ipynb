```python
import pandas as pd
```

```python
csv1 = "Homeless_Data_Merged.csv"
csv2 = "Population by state-revised.csv"
```

```python
df1 = pd.read_csv(csv1)
df1.head()
```

```python
df2 = pd.read_csv(csv2)
df2.head()
```

```python
df2 = df2.drop("State", axis=1)
df2.head()
```

```python
merge_df = pd.merge(df1, df2, how='inner')
merge_df.head()
```

```python
merge_df.to_csv("Dataset.csv")
```
