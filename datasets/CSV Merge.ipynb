```python
import pandas as pd
```

```python
csv1 = "Homeless_Data_Merged.csv"
csv2 = "Population by state-revised.csv"
csv3 = "census_data.csv"
csv4 = "Annual_Homeless_Population.csv"
csv5 = "HUD_Federal_Budge.csv"
```

```python
df1 = pd.read_csv(csv1)
df1.head()
```

```python
df2 = pd.read_csv(csv2)
df2.head()
```

```python
df3 = pd.read_csv(csv3)
df3.head()
```

```python
merge_df = pd.merge(df1, df2, how='inner')
merge_df.head()
```

```python
new_df = pd.merge(merge_df, df3, how='inner')
new_df.head()
```

```python
homeless_rate = ((new_df['homeless'] / new_df['total_population'])*100).round(2)
homeless_rate.head()
```

```python
new_df["homeless_rate"] = homeless_rate
new_df.head()
```

```python
new_df.info()
```

```python
new_df.to_csv("data.csv")
```
